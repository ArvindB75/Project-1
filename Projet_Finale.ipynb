* Q1 : Quels sont les attributs les moins desirables chez les hommes ? Sont elles différentes chez les femmes ?

```python
# Definir les colonnes appartenant à la preference des attributs du partenaire
preference_cols = ['pf_o_att', 'pf_o_sin', 'pf_o_int', 'pf_o_fun', 'pf_o_amb', 'pf_o_sha']

# Creation d'un dataset avec ces colonnes et le genre 
df_q1 = df[['gender'] + preference_cols]

# Eliminer les lignes avec des valeurs manquantes
df_q1 = df_q1.dropna()

# Definir la grille du graphique
plt.figure(figsize=(20, 10))
sns.set(style="whitegrid")

# Faire une boucle de generation de graphes pour chacun des attributs par sexe
for i, col in enumerate(preference_cols, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x='gender', y=col, data=df_q1, palette=['pink', 'blue'])
    plt.title(col)

plt.show()
```

Les hommes accordent généralement plus d'importance à l'attractivité (pf_o_att) que les femmes.
Les femmes accordent plus d'importance à la sincérité (pf_o_sin), à l'intelligence (pf_o_int), à l'ambition (pf_o_amb) et aux intérêts partagés (pf_o_sha) que les hommes.
Les hommes et les femmes accordent à peu près la même importance au sens de l'amusement (pf_o_fun).
les hommes valorisent plus l'attractivité, tandis que les femmes valorisent plus la sincérité, l'intelligence, l'ambition et les intérêts partagés.
Cependant, pour déterminer quels sont les attributs les moins désirables, je dois examiner les attributs qui ont les scores les plus bas pour chaque genre.

```python
# Calculer la preference moyenne pour chaque attribut (hommes et femmes séparemment)
mean_preferences = df_q1.groupby('gender').mean()

# Modifier le DataFrame pour plus de lisibilité
mean_preferences = mean_preferences.transpose()
mean_preferences.reset_index(inplace=True)
melted_preferences = mean_preferences.melt(id_vars='index', var_name='gender', value_name='mean_preference')

# Remplacer le code des genres par le nom des genres
melted_preferences['gender'] = melted_preferences['gender'].replace({0: 'Women', 1: 'Men'})

# Graphe de preferences moyennes de chaque attribut (hommes et femmes séparemment)
plt.figure(figsize=(10, 6))
sns.barplot(x='index', y='mean_preference', hue='gender', data=melted_preferences, palette=['pink', 'blue'])
plt.xlabel('Attribute')
plt.ylabel('Mean Preference Score')
plt.title('Mean Preference Scores for Different Attributes')
plt.legend()
plt.show()

```

Chez les hommes comme pour les femmes les attributs les moins desirables semblent etre l'ambition (pf_O_amb) et les interets partagés (pf_o_sha).

* Q2 : Quelle est l’importance donnée par les participants à l’attribut attractivité dans le choix de selection vs la réalité ?

```python
# Creation d'un DataFrame avec les colonnes de genre et d'attractivité
df_q2 = df[['gender', 'pf_o_att', 'attr_o']]

# Eliminer les lignes avec des valeurs manquantes
df_q2 = df_q2.dropna()

# Afficher les premieres lignes du nouveau dataset
df_q2.head()
```

```python
# Calculer la correlation entre 'pf_o_att' and 'attr_o' pour chaque genre
correlation_women = df_q2[df_q2['gender'] == 0][['pf_o_att', 'attr_o']].corr().iloc[0, 1]
correlation_men = df_q2[df_q2['gender'] == 1][['pf_o_att', 'attr_o']].corr().iloc[0, 1]

correlation_women, correlation_men
```

Ces valeurs sont très proches de zéro, ce qui suggère qu'il n'y a pas de corrélation significative entre ces deux variables, ni pour les hommes ni pour les femmes.

Cela suggère que l'importance que les participants accordent à l'attractivité dans leurs préférences déclarées n'est pas fortement liée à la façon dont leurs partenaires perçoivent leur attractivité. En d'autres termes, ce que les participants pensent vouloir (en termes d'attractivité) n'est pas fortement lié à ce qu'ils obtiennent.

```python
# Creation d'un scatterplot pour la relation entre 'pf_o_att' and 'attr_o' pour chaque genre
plt.figure(figsize=(12, 6))

# Graphe pour les femmes
plt.subplot(1, 2, 1)
sns.scatterplot(x='pf_o_att', y='attr_o', data=df_q2[df_q2['gender'] == 0], alpha=0.5, color='pink')
plt.title('Women')
plt.xlabel('Declared Preference for Attractiveness')
plt.ylabel('Perceived Attractiveness')

# Graphe pour les hommes
plt.subplot(1, 2, 2)
sns.scatterplot(x='pf_o_att', y='attr_o', data=df_q2[df_q2['gender'] == 1], alpha=0.5, color='blue')
plt.title('Men')
plt.xlabel('Declared Preference for Attractiveness')
plt.ylabel('Perceived Attractiveness')

plt.tight_layout()
plt.show()
```

Comme on peut le voir, il n'y a pas de tendance claire ou de corrélation visible entre ces deux variables, ce qui est cohérent avec les coefficients de corrélation calculés précédemment. Cela suggère que l'importance que les participants accordent à l'attractivité dans leurs préférences déclarées ne se reflète pas nécessairement dans la façon dont leurs partenaires perçoivent leur attractivité.

En somme, il semble que ce que les participants pensent vouloir (en termes d'attractivité) n'est pas fortement lié à ce qu'ils obtiennent lors de l'événement de speed dating.

* Q4 : Est ce que les participants peuvent prédire de manière precise leurs valeurs perçue dans le marché ?

```python
# Creation d'un nouveau DataFrame avec les colonnes adequates
df_q4 = df[['gender', 'attr3_1', 'sinc3_1', 'intel3_1', 'fun3_1', 'amb3_1', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o']]

# Eliminer les lignes avec des valeurs manquantes
df_q4 = df_q4.dropna()

# Afficher les premiers resultats du nouveau DataFrame
df_q4.head()
```

```python
# Definition de colonnes des paires d'auto evaluation et d'evaluation par partenaire
evaluation_pairs = [
    ('attr3_1', 'attr_o'),
    ('sinc3_1', 'sinc_o'),
    ('intel3_1', 'intel_o'),
    ('fun3_1', 'fun_o'),
    ('amb3_1', 'amb_o')
]

# Calcul de la correlation pour chaque paire et pour chaque genre
correlations = {}
for pair in evaluation_pairs:
    correlations[pair] = (
        df_q4[df_q4['gender'] == 0][list(pair)].corr().iloc[0, 1],
        df_q4[df_q4['gender'] == 1][list(pair)].corr().iloc[0, 1]
    )

correlations
```

Ces valeurs de corrélation sont faibles, ce qui suggère qu'il n'y a pas de corrélation forte entre les auto-évaluations des participants et les évaluations de leurs partenaires pour chaque attribut. Cela suggère que les participants ne sont pas très précis pour prédire leurs valeurs perçues sur le marché du speed dating.

Cependant, il est intéressant de noter que la corrélation est généralement plus élevée pour les hommes que pour les femmes, ce qui suggère que les hommes peuvent être légèrement plus précis que les femmes pour prédire leurs valeurs perçues.

* Q5 : Pour avoir un 2eme rencard, est il preferable d’être le premier rencard d’un participant ou le dernier dans une session de speed dating ?


```python
# Creation d'un nouveau DataFrame avec les colonnes adequates
df_q5 = df[['gender', 'order', 'dec_o']]

# Eliminer les lignes avec des valeurs manquantes
df_q5 = df_q5.dropna()

# Afficher les premiers resultats du nouveau DataFrame
df_q5.head()
```

```python
# Calcul de la correlation entre 'order' and 'dec_o' pour chaque genre
correlation_order_women = df_q5[df_q5['gender'] == 0][['order', 'dec_o']].corr().iloc[0, 1]
correlation_order_men = df_q5[df_q5['gender'] == 1][['order', 'dec_o']].corr().iloc[0, 1]

correlation_order_women, correlation_order_men
```

La corrélation entre l'ordre du rendez-vous et la décision du partenaire sur un deuxième rendez-vous est de -0.011 pour les femmes et de -0.050 pour les hommes. Ces valeurs sont très proches de zéro, ce qui suggère qu'il n'y a pas de corrélation significative entre ces deux variables, ni pour les hommes ni pour les femmes.

Donc que vous soyez le premier rendez-vous d'un participant ou le dernier lors d'une session de speed dating, cela n'a pas une grande influence sur la décision d'un deuxième rendez-vous.

Cependant, il faut noter que ces valeurs sont négatives, ce qui indique une légère tendance : plus l'ordre du rendez-vous est élevé (c'est-à-dire plus tard dans la session), moins il est probable que le partenaire décide d'un deuxième rendez-vous. Cette tendance est légèrement plus marquée pour les hommes que pour les femmes.

```python
# Creation d'un scatterplot pour la relation entre 'order' and 'dec_o' pour chaque genre
plt.figure(figsize=(12, 6))

# Graphe pour les femmes
plt.subplot(1, 2, 1)
sns.boxplot(x='dec_o', y='order', data=df_q5[df_q5['gender'] == 0], color='pink')
plt.title('Women')
plt.xlabel('Decision on Second Date')
plt.ylabel('Order of Date')

# Graphe pour les hommes
plt.subplot(1, 2, 2)
sns.boxplot(x='dec_o', y='order', data=df_q5[df_q5['gender'] == 1], color='blue')
plt.title('Men')
plt.xlabel('Decision on Second Date')
plt.ylabel('Order of Date')

plt.tight_layout()
plt.show()
```

Comme on peut le voir, il n'y a pas de différence majeure entre les distributions de l'ordre des rendez-vous pour les cas où le partenaire a décidé ou non d'un deuxième rendez-vous, que ce soit pour les hommes ou pour les femmes. Cela confirme l'analyse précédente selon laquelle l'ordre du rendez-vous n'a pas une grande influence sur la décision d'un deuxième rendez-vous.

Cependant, pour les hommes, il est intéressant de noter que la médiane de l'ordre du rendez-vous est légèrement plus élevée pour les cas où le partenaire a décidé de ne pas avoir de deuxième rendez-vous. Cela suggère que les hommes ont légèrement moins de chances d'obtenir un deuxième rendez-vous s'ils sont parmi les derniers rendez-vous de la session. Pour les femmes, cette tendance n'est pas aussi apparente.

En conclusion, bien que l'ordre du rendez-vous n'ait pas une grande influence sur la décision d'un deuxième rendez-vous, il semble y avoir une légère tendance pour les hommes à avoir moins de chances d'obtenir un deuxième rendez-vous s'ils sont parmi les derniers rendez-vous de la session. Pour les femmes, cette tendance n'est pas aussi apparente.

* Q7 : Y a-t-il une différence dans les préférences ou les décisions de match en fonction de l'âge des participants ?

```python
# Creation d'un nouveau DataFrame avec les colonnes adequates
df_q7 = df[['gender', 'age', 'match', 'attr_o', 'sinc_o', 'intel_o', 'fun_o', 'amb_o', 'shar_o']]

# Eliminer les lignes avec des valeurs manquantes
df_q7 = df_q7.dropna()

# Afficher les premiers resultats du nouveau DataFrame
df_q7.head()
```

```python
# Definition des labels d'ages
bins = [18, 25, 30, 35, 40, df_q7['age'].max()]
labels = ['18-25', '26-30', '31-35', '36-40', '41+']

# Creation d'une colonne 'age_group'
df_q7['age_group'] = pd.cut(df_q7['age'], bins=bins, labels=labels)

# Afficher les premiers resultats du nouveau DataFrame
df_q7.head()
```

```python
# Calcul de la moyenne des matchs et des notes d'attributs pour chaque groupe d'ages
average_match_rate = df_q7.groupby('age_group')['match'].mean()
average_attribute_ratings = df_q7.groupby('age_group')[attributes_columns].mean()

average_match_rate, average_attribute_ratings
```

```python
# Visualisation de la moyenne de match pour chaque groupe d'ages
average_match_rate.plot(kind='bar', figsize=(12, 6), color='blue', alpha=0.7)
plt.title('Average Match Rate by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Match Rate')
plt.show()

# Visualisation des notes d'attributs pour chaque groupe d'ages
average_attribute_ratings.plot(kind='bar', figsize=(15, 6))
plt.title('Average Attribute Ratings by Age Group')
plt.xlabel('Age Group')
plt.ylabel('Average Attribute Rating')
plt.legend(title='Attribute')
plt.show()
```

Taux de match moyen pour chaque groupe d'âge :

Les participants les plus jeunes (18-25 ans) et les plus âgés (plus de 40 ans) ont tendance à avoir un taux de match légèrement plus élevé que les autres groupes d'âge. Les participants âgés de 36 à 40 ans ont le taux de match le plus bas.

Evaluations moyennes pour chaque attribut, pour chaque groupe d'âge :

Attractivité (attr_o) : Les participants les plus jeunes et les plus âgés sont légèrement mieux notés en termes d'attractivité.
Sincérité (sinc_o) : Les évaluations de sincérité sont assez similaires pour tous les groupes d'âge.
Intelligence (intel_o) : Les participants âgés de 26 à 35 ans sont légèrement mieux notés en termes d'intelligence.
Sens de l'amusement (fun_o) : Les participants les plus jeunes et les plus âgés sont légèrement mieux notés en termes de sens de l'amusement.
Ambition (amb_o) : Les participants âgés de 26 à 40 ans sont légèrement mieux notés en termes d'ambition.
Intérêts partagés (shar_o) : Les participants les plus jeunes et les plus âgés ont légèrement de meilleurs scores en termes d'intérêts partagés.
Ces résultats suggèrent que l'âge des participants peut avoir une influence sur leurs préférences et leurs décisions de match lors d'un événement de speed dating.

* Q10 : Comment les intérêts des participants (par exemple, les films, la musique, le shopping, etc.) influencent-ils leurs décisions de match ?

```python
# Creation d'un nouveau DataFrame avec les colonnes adequates
interests_columns = ['sports', 'tvsports', 'exercise', 'dining', 'museums', 'art', 'hiking', 'gaming', 'clubbing', 'reading', 'tv', 'theater', 'movies', 'concerts', 'music', 'shopping', 'yoga']
df_q10 = df[['gender', 'match'] + interests_columns]

# Eliminer les lignes avec des valeurs manquantes
df_q10 = df_q10.dropna()

# Afficher les premiers resultats du nouveau DataFrame
df_q10.head()
```

```python
# Calcul de la correlation entre activités et 'match' pour chaque genre
correlations_interests_match = {}
for column in interests_columns:
    correlations_interests_match[column] = (
        df_q10[[column, 'match']].corr().iloc[0, 1],
        df_q10[df_q10['gender'] == 0][[column, 'match']].corr().iloc[0, 1],
        df_q10[df_q10['gender'] == 1][[column, 'match']].corr().iloc[0, 1]
    )

correlations_interests_match
```

```python
# Conversion des dictionnaires de correlations en DataFrame
correlations_df = pd.DataFrame(correlations_interests_match, index=['All', 'Women', 'Men']).T

# Creation d'un barplot de correlations
correlations_df.plot(kind='barh', figsize=(12, 6), alpha=0.7)
plt.title('Correlation between Interests and Match Decision')
plt.xlabel('Correlation')
plt.ylabel('Interest')
plt.legend(title='Gender')
plt.gca().invert_yaxis()
plt.show()
```

On observe que certaines activités, comme le clubbing, le yoga, et l'art, ont une corrélation légèrement positive avec la décision de match pour tous les participants. Par ailleurs, pour les femmes, les musées, l'art, et les concerts ont une corrélation légèrement positive avec la décision de match. Pour les hommes, les sports, le dîner, et l'exercice ont une corrélation légèrement positive avec la décision de match.

Ces résultats suggèrent que, bien que l'influence des intérêts sur les décisions de match soit limitée, certaines activités peuvent avoir une corrélation légèrement plus forte sur la décision d'un match.

* Q11 : Y a-t-il une différence dans les préférences ou les décisions de match en fonction de l'objectif des participants lors de l'événement de speed dating (par exemple, s'amuser, rencontrer de nouvelles personnes, obtenir un rendez-vous, etc.) ?

```python
# Mapping entre 'goal' et les noms correspondant dans la documentation
goal_dict = {1: 'Seemed like a fun night out', 2: 'To meet new people', 
             3: 'To get a date', 4: 'Looking for a serious relationship', 
             5: 'To say I did it', 6: 'Other'}

# Creation d'une colonne pour le nom des 'goal'
df['goal_name'] = df['goal'].map(goal_dict)

# Calcul des moyennes de matchs pour chaque 'goal'
average_match_rate_goal = df.groupby('goal_name')['match'].mean().sort_values(ascending=False)

# Visualisation des moyenne de matchs en fonction des 'goal'
plt.figure(figsize=(12,6))
average_match_rate_goal.plot(kind='barh', color='blue', alpha=0.7)
plt.title('Average Match Rate by Dating Goal')
plt.xlabel('Average Match Rate')
plt.ylabel('Dating Goal')
plt.gca().invert_yaxis()
plt.show()
```

Il semble que ceux qui cherchent une relation sérieuse, qui sont la pour s'amuser ou qui veulent simplement dire qu'ils l'ont fait ont un taux de correspondance légèrement plus élevé que ceux qui sont la pour rencontrer de nouvelles personnes ou obtenir un rendez-vous. Cependant, les différences ne sont pas très importantes et peuvent être dues à des variations aléatoires.
